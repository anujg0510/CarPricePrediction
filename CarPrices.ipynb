```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.2)
```

```python
X_train.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
import numpy as np
rf_random = RandomForestRegressor()
#RandomForestRegressor() needs a lot of parameters
#We'll used RandomizedSearchCV below to find out the optimal value of each of these parameters
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
###Hyperparameters
###RandomizedSearchCV

#No of trees in the random forest
n_estimators = [int(x) for x in np.linspace(start=100, stop=1200, num=12) ]

#No of features to consider at every split
max_features = ["auto", "sqrt"]

#Max number of levels in the tree
max_depth = [int(x) for x in np.linspace(5,30, num=6)]

#Min number of samples required to split a node
min_sampe_split = [2,5,20,15,100]

#Min number of samples required at each leaf node
min_samples_leaf = [1,2,5,10]
```

```python
#Create the Random Grid

random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_sampe_split,
               'min_samples_leaf': min_samples_leaf
              }
print(random_grid)
```

```python
rf = RandomForestRegressor()
```

```python
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, n_iter=10, scoring="neg_mean_squared_error", cv=5, n_jobs=1, verbose=2, random_state=42)
```

```python
rf_random.fit(X_train, y_train)
```

```python
predictions = rf_random.predict(X_test)
print(predictions)
```

```python
sns.distplot(y_test-predictions)
```

```python
plt.scatter(y_test, predictions)
```
